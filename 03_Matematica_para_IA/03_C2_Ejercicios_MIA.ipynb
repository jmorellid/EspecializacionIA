### Ej 2

¿Cómo calcular ́ıa la recta y = a1x+a0 que mejor aproxima (en t ́erminos de la distancia 2) los datos obtenidos?

(x0, y0) ... (xn, yn)

Se utiliza la expresión $y = a_1 x + a_0$ para aproximar a y. En caso que y no sea combinación lineal de la expresión $a_1 x + a_0$, se debe encontrar el conjunto de valores [a1, a0] que mejor aproxima a y.

Quedaría entonces la expresión:

$\hat y = a_1 x + a_0$

Como hay numerosos valores de x e i, expreso el conjunto de ecuaciones matricialmente:

- Y = vector de valores yi

- A = matriz de valores [xi, 1]

- z = [a0, a1]

$$\hat Y = Az$$

Para encontrar el conjunto de z que mejor aproximan a y, debo definir la manera de medir distancia. Para ello defino como distancia a la recta a la norma l2 del error. Entonces debo minimizar:

$$Loss = min ||Y - \hat Y|| => min ||Y - A. z||$$

Y el conjunto de parametros z optimos serán:
z0 = argmin ||Y - A. z||

Para minimizar esa distancia utilizaré el producto interno asociado a la norma L2.

El problema planteado consiste entocnes en un ejercicio de proyección ortogonal sobre el subespacio conformado por los vectores de la matriz A.


Para encontrar los valores de z óptimos debo encotnrar los valores de z para los cuales Az pertenece a S, generado por las columnas de A.

Tengo que proyectar Y en el subespacio generado por los vecotres de A. Sabemos entonces que Az pertenece a S. Por lo tanto, Y - Az proyectado sobre s debe ser 0 para minimizar el error. Esto se debe a que el vector ortogonal a S perteneciente a Y minimizará la distancia de Y a S y, consecuentemente, el error de la medición.

$$<A, Y - A.z> = 0$$


$$(A^t(Y - A.z) = 0$$


$$(A^tY - A^tA.z) = 0$$


$$A^tY = A^tA.z$$


$$(A^tA)^{-1}A^tY = z$$


z será el valor que mejor aproxima los puntos con una recta.

### Ejercicio 3

1. A = $
\begin{vmatrix}
1 & 2\\
3 & 4\\
5 & 6\\
\end{vmatrix}
$

2. B = $
\begin{vmatrix}
1 & 2\\
1 & 2\\
\end{vmatrix}$

Según lo deducido del método de ortogonalización de Graham Schmidt:

$$v1 = g1$$
$$v2 = \frac{<v2,g1>}{||g1||^2} + g2  $$

Dicha deducción se utiliza para reconstruir los vectores mediante la facotrización matricial siguiente:

$$ Q =
\begin{vmatrix}
\frac{g1}{||g1||}, \frac{g2}{||g2||}
\end{vmatrix}$$


$$ R =
\begin{vmatrix}
||g1|| & \frac{<v2,g1>}{||g1||}\\
0 & ||g2|| \\
\end{vmatrix}$$

Se agregó el término de la norma de tal manera de normalizar la matriz Q y así generar una matriz ORTONORMAL.


```python
import numpy as np
A = np.array([[1, 2], [3, 4], [5, 6]])

B = np.array([[1, 2], [1, 2]])
```

```python
v1 = A.T[0].reshape(3,1)
v2 = A.T[1].reshape(3,1)

#Ortogonalización de vectores
g1 = v1
g2 = v2 - np.dot(v2.T, v1)/np.dot(v1.T ,v1)*  v1

#prueba de ortogonalidad
print('¿Los vectores son ortogonales?', np.dot(g2.T, g1)[0][0] < 10**-10 )

#cálculo de normas de vectores ortogonales
g1_norm = (np.dot(g1.T,g1)**0.5)[0][0]
g2_norm = (np.dot(g2.T,g2)**0.5)[0][0]

#normalización de vecores ortogonales
g1_unit = g1 / g1_norm
g2_unit = g2 / g2_norm

#armado de matriz Q
Q = np.hstack((g1_unit, g2_unit))

# Armado de matriz R

R = np.array([[g1_norm, np.dot(v2.T, g1)[0][0]/g1_norm ],
              [0 , g2_norm]])

print('la matriz factorizada según QR A:\n', np.matmul(Q, R))

# Creación de función que realiza la descomposición QR para hacer ej b

def qr_descomposition(A):
    
    filas = A.shape[0]
    columnas = A.shape[1]
    
    matriz_ortogonal = np.zeros(shape=[filas,columnas])
    matriz_ortonormal = np.zeros(shape=[filas,columnas])
    R_matrix = np.zeros(shape=[columnas, columnas])

    for i in range(columnas):
        matriz_ortogonal[:,i] = A[:,i]
        g_j = 0

        for j in range(i):
            g_j = matriz_ortogonal[:,j]
            v_projection_gj = np.dot(A[:,i].T, g_j)/ np.dot(g_j.T ,g_j)
            matriz_ortogonal[:,i] = matriz_ortogonal[:,i] - v_projection_gj * g_j
            R_matrix[i,j] = v_projection_gj *  np.dot(g_j.T ,g_j)**0.5

        R_matrix[i,i] = np.dot(matriz_ortogonal[:,i], matriz_ortogonal[:,i])**0.5

        matriz_ortonormal[:,i] = matriz_ortogonal[:,i] / R_matrix[i,i]
    
    return matriz_ortonormal, R_matrix.T

Qb, Rb = qr_descomposition(B)

# como la proyeccion de V2 sobre 
# V1 cae totalmente dentro de V1,
# el segundo vector se anula.. Y pierde relevancia 
# El algoritmo arroja nans, los reemplazo por valores cualesquiera.
Qb[:,1] = [1,1]

print('la matriz factorizada según QR B:\n', np.matmul(Qb, Rb))

```
