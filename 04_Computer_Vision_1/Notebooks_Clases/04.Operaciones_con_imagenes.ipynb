# Operaciones en imágenes

Veremos operaciones ariméticas y lógicas

```python
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
# Cargo las imágenes de ejemplo
img1 = cv.imread('imgBloque1.bmp',cv.IMREAD_GRAYSCALE)
img2 = cv.imread('imgBloque2.bmp',cv.IMREAD_GRAYSCALE)

plt.imshow(img1,cmap='gray')
plt.show()

plt.imshow(img2,cmap='gray')
plt.show()
```

```python
# Suma de imágenes
img_suma = img1 + img2
plt.imshow(img_suma,cmap='gray')

print(img1[100, 280], img2[100,280])
```

```python
# Suma de imágenes
img_suma = cv.add(img1,img2)

plt.imshow(img_suma,cmap='gray')
plt.show()
img_suma[100,280]
# Explicar la diferencia !

"""
La suma directa entre img1 + img2 arroja valores que sobrepasan el límite de la escala de grises (255). Para aquellos valores la imagen "Se pasa" y se decodifica como img1 + img2 - 255
Lo que arroja colores cercanos al negro.

En cv.add en cambio, el valor máximo es 255 y la sumatoria es acotada por ese valor.
"""
```

```python
# Sumar con distinto peso
alpha = 0.2
img_suma = cv.addWeighted(img1,alpha,img2,1-alpha,0)

plt.imshow(img_suma,cmap='gray')
plt.show()

# Respetando la ecuación propuesta, dará siempre el mismo resultado si usamos la suma de Numpy ?
#aa = img1*alpha+img2*(1-alpha)
#plt.imshow(aa,cmap='gray')
#plt.show()

"""
No por lo concluido anteriormente.
"""
```

####  Ejercicio:
A partir de 2 imágenes crear una transición suave entre ellas

```python
# Operaciones lógicas

# Tomo los ciclos de clock al iniciar
e1 = cv.getTickCount()

img2 = np.zeros(img1.shape, dtype=np.uint8)
img2[100:200,100:450] = 255 #0xFF

img3 = cv.bitwise_and(img1,img2)

# Tomo los ciclos de clock al finalizar
e2 = cv.getTickCount()
time = (e2 - e1)/ cv.getTickFrequency()
print('Se tardaron {:0.4} milisegundos en relizar la operación'.format(time*1000))

plt.imshow(img3,cmap='gray')
plt.show()

img3 = cv.bitwise_or(img1,img2)
plt.imshow(img3,cmap='gray')
plt.show()
```
