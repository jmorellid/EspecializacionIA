## Importar librer√≠as

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import os
 
import numpy as np

import cv2 as cv
```

```python
images_list = ['./Imagenes/' +img for img in os.listdir(path='./Imagenes')]
print(images_list)

imagenes_contraste = images_list[:3]
imagenes_whitepatch = images_list[3:]
```

```python
## Algoritmo coordenadas crom√°icas

for i, imagen_contrastosa in enumerate(imagenes_contraste):
    
    amt_images = len(imagenes_contraste)
    img_idx_1 = int(str(amt_images)+str(2)+str(i*2+1))
    img_idx_2 = int(str(amt_images)+str(2)+str(i*2+2))
    
    IC = cv.imread(imagen_contrastosa)
    IC = cv.cvtColor(IC,cv.COLOR_BGR2RGB)
    IC = np.array(IC)
    
    sum_matrix = 1/np.sum(IC, axis=2)
    chromatic_image = IC * sum_matrix[:,:,None]    
    
    plt.figure(3, figsize=(10,10))
    
    plt.subplot(img_idx_1), plt.imshow(IC)    
    plt.subplot(img_idx_2), plt.imshow(chromatic_image)
    plt.show()
```

```python
## Implementaci√≥n white patch

fig, ax = plt.subplots(nrows=len(imagenes_whitepatch), ncols=4, figsize=(20, 20))

for i, imagen_white in enumerate(imagenes_whitepatch):
    
    IC = cv.imread(imagen_white)
    IC = cv.cvtColor(IC,cv.COLOR_BGR2RGB)
    IC = np.array(IC)
    
    R_max = np.max(IC[:,:,0])
    G_max = np.max(IC[:,:,1])
    B_max = np.max(IC[:,:,2])
    
    max_vector = np.array([R_max, G_max, B_max])
    
    whitepatch_image = (IC / max_vector * 255).astype(int) 

    ax[i,0].imshow(IC)    
    ax[i,1].imshow(whitepatch_image)
    
    ax[i,2].hist(IC[:,:,0].flatten(), bins=50, alpha=0.4, color='r', histtype='step')
    ax[i,2].hist(IC[:,:,1].flatten(), bins=50, alpha=0.4, color='g', histtype='step')
    ax[i,2].hist(IC[:,:,2].flatten(), bins=50, alpha=0.4, color='b', histtype='step')
    
    ax[i,3].hist(whitepatch_image[:,:,0].flatten(), bins=50, alpha=0.4, color='r', histtype='step')
    ax[i,3].hist(whitepatch_image[:,:,1].flatten(), bins=50, alpha=0.4, color='g', histtype='step')
    ax[i,3].hist(whitepatch_image[:,:,2].flatten(), bins=50, alpha=0.4, color='b', histtype='step')
plt.show()
```

```python
# Save the Tito: Try and get a cleaner image.

IC = cv.imread(imagenes_whitepatch[3])
IC = cv.cvtColor(IC,cv.COLOR_BGR2RGB)
IC = np.array(IC)

IC[:,:,0] = cv.add(IC[:,:,0], +20) * 2
IC[:,:,1] = cv.add(IC[:,:,1], -20) / 2
IC[:,:,2] = cv.add(IC[:,:,2], -70) / 2

R_max = np.max(IC[:,:,0])
G_max = np.max(IC[:,:,1])
B_max = np.max(IC[:,:,2])

max_vector = np.array([R_max, G_max, B_max])

whitepatch_image = (IC / max_vector * 255).astype(int) 

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20, 10))

ax[0].imshow(IC)    
ax[1].imshow(whitepatch_image)

ax[2].hist(IC[:,:,0].flatten(), bins=50, alpha=0.4, color='r', histtype='step')
ax[2].hist(IC[:,:,1].flatten(), bins=50, alpha=0.4, color='g', histtype='step')
ax[2].hist(IC[:,:,2].flatten(), bins=50, alpha=0.4, color='b', histtype='step')

ax[3].hist(whitepatch_image[:,:,0].flatten(), bins=50, alpha=0.4, color='r', histtype='step')
ax[3].hist(whitepatch_image[:,:,1].flatten(), bins=50, alpha=0.4, color='g', histtype='step')
ax[3].hist(whitepatch_image[:,:,2].flatten(), bins=50, alpha=0.4, color='b', histtype='step')
plt.show()

# Robotito Saved
```
